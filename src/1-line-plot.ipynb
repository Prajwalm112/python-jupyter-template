### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np

# Define x values ranging from 0 to 10
x = np.arange(0, 11, 1)

# Calculate y values as the square of x
y = x**2

# Create the line plot
plt.plot(x, y, marker='o', linestyle='-', color='b', label='y = x^2')

# Add a title and labels for both axes
plt.title("Line Plot of y = x²")
plt.xlabel("x values")
plt.ylabel("y values")

# Add a grid to the plot
plt.grid(True, linestyle='--', alpha=0.6)

# Show the legend
plt.legend()

# Display the plot
plt.show()
```
